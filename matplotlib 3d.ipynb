```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
theta = np.linspace(-4 * np.pi, 4 * np.pi, 100)
z = theta * 0.1
x = np.sin(theta)
y = np.cos(theta)
```

```python
fig = plt.figure(figsize=(15,15))
ax = fig.add_subplot(1,1,1,projection='3d')
ax.view_init(elev = 30, azim = 30)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
ax.dist = 10
#ax.set_aspect('equal')
ax.axis("equal")

ax.set_xlim3d([-5, 5])
ax.set_ylim3d([-5, 5])
ax.set_zlim3d([-5, 5])

#ax.set_xticklabels([])
ax.plot(x, y, z)
plt.show()
```
